## Data loading

```python
import wget
import pandas as pd
import numpy as np 
import os
from tqdm.notebook import tqdm
```

Parsing the start and the end of the burst

```python
url = 'https://swift.gsfc.nasa.gov/results/batgrbcat/summary_cflux/summary_general_info/summary_burst_durations.txt'
output_directory = '../data/'
filename = wget.download(url, out=output_directory)
filename
```

```python
burst_durations_t100 = {}

with open(filename, "r") as input_file:
    for line in input_file:
        if not line.startswith('#'):
            try:
                burst_name = [i.strip() for i in line.split('|')][:-1][0]
                burst_trig = [i.strip() for i in line.split('|')][:-1][1]
                burst_strat = float([i.strip() for i in line.split('|')][:-1][3])
                burst_end = float([i.strip() for i in line.split('|')][:-1][4])
                
                

                burst_durations_t100[burst_name] = {}
                
                burst_durations_t100[burst_name]['strat'] = burst_strat
                burst_durations_t100[burst_name]['end'] = burst_end
                burst_durations_t100[burst_name]['trig'] = burst_trig
               

            except :
                print(line)
```

```python
burst_durations_t100
```

```python
burst_durations_t100_df = pd.DataFrame(burst_durations_t100).T.reset_index() 
burst_durations_t100_df.columns = ['burst_name', 'strat', 'end', 'trig']
burst_durations_t100_df.to_csv('burst_durations_t100_df.csv', index = False)
```

```python
burst_names = list(burst_durations_t100.keys())
```

```python
len(burst_names)
```

```python
for i in burst_names:
    if len(burst_durations_t100[i]['trig']) == 6:
        burst_durations_t100[i]['trig'] = '00' + burst_durations_t100[i]['trig'] + '000'
```

```python
# t90 duration
burst_durations_t90 = {}

with open(filename, "r") as input_file:
    for line in input_file:
        if not line.startswith('#'):
            try:
                burst_name = [i.strip() for i in line.split('|')][:-1][0]
                burst_trig = [i.strip() for i in line.split('|')][:-1][1]
                burst_strat = float([i.strip() for i in line.split('|')][:-1][5])
                burst_end = float([i.strip() for i in line.split('|')][:-1][6])
                
                

                burst_durations_t90[burst_name] = {}
                
                burst_durations_t90[burst_name]['strat'] = burst_strat
                burst_durations_t90[burst_name]['end'] = burst_end
                burst_durations_t90[burst_name]['trig'] = burst_trig
                burst_durations_t90[burst_name]['duration'] = burst_end - burst_strat
               

            except :
                print(line)
```

```python
pd.DataFrame(burst_durations_t90).T.reset_index().to_csv('t90_duration.csv', index = False)
```

download and preprocess bursts data (64ms)

```python
for grb_name in tqdm(burst_names):
    try:
        trig = burst_durations_t100[grb_name]['trig']
        url = f'https://swift.gsfc.nasa.gov/results/batgrbcat/{grb_name}/data_product/{trig}-results/lc/64ms_lc_ascii.dat'
        output_directory = '../data/raw_data/'
        filename = wget.download(url, out=output_directory, bar = None)
        
        df = pd.read_csv( filename, sep = ' ', header=None)
        df = df[[0,1,3,5,7,9]]
        df.columns = ['center', '15_25_keV', '25_50_keV', '50_100_keV', '100_350_keV', '15_350_keV']
        df = df[(df.center >= burst_durations_t100[grb_name]['strat']) & (df.center <= burst_durations_t100[grb_name]['end'])].reset_index(drop=True)
        
        df.to_feather(f'../data/raw_data/{grb_name}.fth')
        os.remove(filename)
        
    except:
        print(grb_name + ' failed')
```
